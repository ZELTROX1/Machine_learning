Plot the two graphs on same plot.The graph between the two given integer array X and Y and the graph between two integer array X and Y where X should hold 20 even points starting from 0 and y should be equal to X where y[i] = 2*x[i]. Plot should have x label and y label and the legend to differentiate between the two graphs.

Consider,

x1 =[0,1,2,3,4,5,6,7,9,10], y1=[1,2,5,10,17,26,37,50,82,101]
x2=np.arange(0,40,2), y2=2*x2
Find the value x (in the range of 10) of both the plots when the value of y is 60. For each plot print the output to new line.

Output Format: lower_limit to upper_limit

```python
x1 =[0,1,2,3,4,5,6,7,9,10]
y1=[1,2,5,10,17,26,37,50,82,101]
x2=np.arange(0,40,2)
y2=2*x2
plt.plot(x1,y1,'--o',label='Line1')
plt.plot(x2,y2,'-o',label ='Line2')
plt.grid()
plt.legend()
x_value_custom_data = next(x for x, y in zip(x1, y1) if y >= 60)
x_value_generated_data = next(x for x, y in zip(x2, y2) if y >= 60)

# Printing the output
print(f"{x_value_custom_data} to {x_value_custom_data + 1}")
print(f"{x_value_generated_data} to {x_value_generated_data + 1}")
plt.show()
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
colors = ['black', 'red', 'violet', 'blue', 'yellow', 'green'] #array of differnet colors

plt.scatter(year, salary, s = population, c = colors) #s=population will cause each data point to have size with respect to the value stored in population at that particular index.
#year[0] vs salary[0] with size=population[0] and color=colors[0]
plt.title('Bubble Chart')
plt.show()
```

```python
c = np.arange(0,36,6)

plt.scatter(year, salary, s=population, marker='^', c=c, alpha=0.4, edgecolor='blue')
plt.show()
```
